# Are there any missing values

```python
df.isna().sum()
```

```python
df['retail_price'].fillna(df['retail_price'].median(), inplace=True)
df['discounted_price'].fillna(df['discounted_price'].median(), inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['uniq_id'],inplace = True)
```

```python
df.isna().sum()
```

## Calculating  the discount percentage

Discount Percentage

```python
x = df['retail_price'] - df['discounted_price']
y = (x/df['retail_price']) * 100
df['discount_percentage'] = y
```

```python
df.head()
```

## Changing of time into proper format

Changing of crawl_timestamp into time and date proper format

```python
df['crawl_timestamp'] = pd.to_datetime(df['crawl_timestamp'], errors='coerce')
```

```python
df.head()
```

# Univariate Analysis

### Bar chart

```python
premium_products = df['is_FK_Advantage_product'].value_counts()
premium_products
```

```python
plt.style.use('dark_background')
plt.figure(figsize = (8,6))
premium_products.plot(kind = 'bar', color=['skyblue', 'salmon'])
plt.title('Bar chart representing about FK_Advantage_product')
plt.xlabel('FK_Advantage_product')
plt.ylabel('Frequency')
plt.xticks(rotation = 0)
plt.show()
```

## Box plot

```python
plt.figure(figsize=(13, 6))
sns.boxplot(df)
```

```python
plt.figure(figsize=(12, 6))
top_brands = df['brand'].value_counts().nlargest(10).index  # Top 10 brands
sns.countplot(data=df[df['brand'].isin(top_brands)], x='brand', order=top_brands)
plt.title('Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Top 10 Products by Discounted Price

```python
top_products = (df.groupby('product_name')['discounted_price'].sum().nlargest(10))
top_products.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Top 10 Products by Discounted Price')
plt.ylabel('')
plt.show()
```

## Top 10 Brands by Discounted Price

```python
top_brands = (df.groupby('brand')['discounted_price'].sum().nlargest(10))
top_brands.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Top 10 Brands by Discounted Price')
plt.ylabel('')
plt.show()
```

## Top 10 Brands with Highest Average Discount Percentage

```python
top_brands = df.groupby('brand')['discount_percentage'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(20, 5))
sns.barplot(x=top_brands.index, y=top_brands.values, palette='rocket')
plt.xticks(rotation=45)
plt.title('Top 10 Brands by Average Discount Percentage')
plt.ylabel('Average Discount (%)')
plt.show()
```

# Bivariate Analysis

# Top 10 Products with Highest Discounted Price

```python
discount_by_product = df.groupby('product_name')['discounted_price'].max().nlargest(10)
discount_by_product.plot(kind='bar', figsize=(20, 5), color='purple')
plt.title('Top 10 Products with Highest Discounted Price')
plt.xticks(rotation = 90)
plt.xlabel('Product Name')
plt.ylabel('Discounted Price')
plt.xticks(rotation=45)
plt.show()
```

## Line plot of Retail Price over time

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='crawl_timestamp', y='retail_price', label='Retail Price')
plt.title('Retail Price Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Retail Price')
plt.xticks(rotation=45)

plt.show()
```

## Line plot of Dicounted Price over time

## how retail_price,product_rating,overall_rating related to each other 

```python
df['product_rating'] = pd.to_numeric(df['product_rating'], errors='coerce')
df['overall_rating'] = pd.to_numeric(df['overall_rating'], errors='coerce')
cleaned_df = df[['retail_price', 'product_rating', 'overall_rating']].dropna()
rating_corr_daily_use_products = cleaned_df.corr()
print(rating_corr_daily_use_products)
```

```python
sns.heatmap(rating_corr_daily_use_products,vmin=-1,vmax=1,annot=True)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='retail_price', y='discounted_price', hue='is_FK_Advantage_product', alpha=0.7)
plt.title('Retail vs Discounted Price by FK Advantage Product')
plt.xlabel('Retail Price')
plt.ylabel('Discounted Price')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='retail_price', fill=True, color='skyblue', bw_adjust=0.5)
plt.title('Density Plot of Retail Price')
plt.xlabel('Retail Price')
plt.ylabel('Density')
plt.show()
```
